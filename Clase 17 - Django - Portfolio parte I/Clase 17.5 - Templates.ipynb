# Plantillas / Templates
##### Son archivos que nos permiten separar la vista de la estética, es decir, guardar en un archivo separado de todo lo que guardábamos en "documento", para así enviar por la HttpsResponse

## Pasos para crear nuestro Template
#### - Dentro de nuestro proyecto creamos una carpeta que se puede llamar "templates" o "plantillas"  
#### - Creamos dentro de esta un archivo .html que será nuestro template; ejemplo "template1.html"

#### Dentro de nuestro template podemos crear nuestra página HTML; si ponemos el signo "!" y tabulamos, VSC nos completará con el esqueleto básico html



## Utilizar nuestro Template
#### Para que nuestro template sea accesible, tenemos que crear la función correspondiente en nuestra "View"
##### Importamos la clase "Template" de Django


##### Abrimos nuestro template creado (archivo.html) y creamos una instancia "template" con su contenido


##### Importamos también la clase "Context"; esto es porque los objetos "template" tienen un método "render" para renderizar nuestro html: esto necesita un objeto "Context" como parámetro


#### El contexto puede estar vacío pero también nos sirve para pasarle valores a nuestro html creando un diccionario y que en nuetro html invocaremos llamando a su clave {{clave}} para obtener el valor



##### El resultado del template renderizado es lo que retornaremos


##### Solo nos resta agregar la función a nuestras URLs
