### Que es un Framework?
##### Es un entorno de trabajo, un ecosistema que nos facilita crear algo, en este caso un sitio web, sin mayores esfuerzos


## MODEL - MODELO
#### Es quien almacena y administra los datos
#### Por lo general cuando hablamos de "model" hablamos de una base de datos y su respectiva capa de acceso a datos

## VIEW - VISTA
#### Es la interfaz que interacta con el usuario, lo que el usuario ve
#### Contiene todas las funcionalidades que interacten con el usuario y contiene botones, cajas de textos, imgenes, etc.
#### Es una representacin visual de los datos y puede tomar las formas que queramos; un grfico, un formulario, etc

## Controller - Controlador
#### Es quien conecta el modelo y la vista segn sea necesario
#### Es una gran funcin que interpreta las acciones y solicitudes del usuario, aplica la lgica de negocio para saber que hacer, interacta con el modelo para obtener y enviarle datos y luego los aplica en la view para mostrrselos al usuario



#### En un caso prctico de uso podemos ver como el usuario ingresa a una pgina a travs de su cliente web o navegador. Esto hace que la vista envi al controlador la solicitud de los datos, este haga su lgica consultando en el modelo y devolviendo a la vista los datos para que puedan ser presentados
#### Algo bueno de este modelo, como podemos observar en la imagen, es que el controlador es independiente del tipo de dispositivo; en este caso el controlador har el mismo trabajo y luego la vista ser quien se acomode para mostrarse de diferentes formas y tamaos dependiendo de qu dispositivo sea el que est haciendo el request
