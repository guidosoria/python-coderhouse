# Django
##### Para realizar un proyecto web necesitaremos usar un __framework__ que nos facilite la tarea; el más sencillo de ellos es Flask, pero nosotros utilizaremos __Django__ que es más completo y de código abierto
### ¿Que es un Framework?
##### Es un entorno de trabajo, un ecosistema que nos facilita crear algo, en este caso un sitio web, sin mayores esfuerzos


## MODEL - MODELO
#### Es quien almacena y administra los datos
#### Por lo general cuando hablamos de "model" hablamos de una base de datos y su respectiva capa de acceso a datos

## VIEW - VISTA
#### Es la interfaz que interactúa con el usuario, lo que el usuario ve
#### Contiene todas las funcionalidades que interactúen con el usuario y contiene botones, cajas de textos, imágenes, etc.
#### Es una representación visual de los datos y puede tomar las formas que queramos; un gráfico, un formulario, etc

## Controller - Controlador
#### Es quien conecta el modelo y la vista según sea necesario
#### Es una gran función que interpreta las acciones y solicitudes del usuario, aplica la lógica de negocio para saber que hacer, interactúa con el modelo para obtener y enviarle datos y luego los aplica en la view para mostrárselos al usuario



#### En un caso práctico de uso podemos ver como el usuario ingresa a una página a través de su cliente web o navegador. Esto hace que la vista envié al controlador la solicitud de los datos, este haga su lógica consultando en el modelo y devolviendo a la vista los datos para que puedan ser presentados
#### Algo bueno de este modelo, como podemos observar en la imagen, es que el controlador es independiente del tipo de dispositivo; en este caso el controlador hará el mismo trabajo y luego la vista será quien se acomode para mostrarse de diferentes formas y tamaños dependiendo de qué dispositivo sea el que está haciendo el request

## DJANGO y MVC
#### En el caso de Django, si bien utiliza MVC, hace unos cambios de nombres

#### En el caso de Django, se hacen un par de cambios de nombre es cuanto a lo que es MVC


#### MODEL in MVC == MODEL in Django
#### VIEW in MVC == TEMPLATE in Django
#### CONTROLLER in MVC == VIEW in Django

### EN Django entonces
#### Cuando hablemos de "Modelo" estaremos hablando de la "base de datos"
#### Cuando hablemos de "Controladores" estaremos hablando de la "funciones de Python"
#### Cuando hablemos de "Template" estaremos hablando de código HTML

#### Podemos decir que Django utiliza un modelo MTV: Model - Template - View
