# HTML HyperText Markup Language
#### Es el lenguaje estándar para crear y estructurar páginas we
#### Utiliza una serie de etiquetas y elementos para definir el contenido y la estructura de la página, como encabezados, párrafos, imágenes, enlaces y otros componentes multimedia
#### HTML no es un lenguaje de programación, sino un lenguaje de marcado que describe cómo se deben mostrar los elementos en el navegador web

# Anatomía de un elemento HTML

## Partes principales


#### __Etiqueta de Apertura__
#### Consiste en el nombre del elemeto, encerrado por paréntesis angulares < > de apertura y cierre. Establece dónde comienza a tener efecto el elemento

#### __Etiqueta de Cierre__
#### Es igual que la de apertura excepto que incluye una barra de cierre / antes del nombre de la etiqueta. Establece dónde termina el elemento

#### __Contenido__
#### Es el contenido del elemento

#### __El elemento__
#### La etiqueta de apertura, cierre más el contenido conforman el elemento

## Atributos


#### Los atributos contienen información adicional acerca del elemento, la cual no quieres que aparezca en el contenido real del elemento
#### En el ejemplo de arriba "class" es el nombre del atributo y "editos-note" el valor del mismo
#### En este caso por ejemplo, el atributo "class" permite darle al elemento un nombre identificativo, que se puede utilizar luego para apuntarle al elemento información de estilo y demás cosas

## Anidar elementos


#### También podemos colocar elemntos dentro de otros elementos: a esto se le llama __anidamiento__

## Elementos vacíos


#### Algunos elementos no poseen contenido y son llamados __elementos vacíos__

# Elementos


### \<!DOCTYPE html> - Tipo de Documento
#### Es un preámbulo requerido en HTML
### \<html> \</html> - Elemento HTML
#### Encierra todo el contenido de la página entera; también conocido como elemento raíz o root element
### \<head> \</head> - Elemento HEAD
#### Actúa como un contenedor de todo aquello que queremos incluir en la página HTML que no es contenido visible por los visitantes de la página. Incluye cosas como "keyword", utilizados para los resultados de búsqueda, código CSS, etc
### \<meta charset="utf-8">
####  Establece el juego de caracteres que tu documento utilizará; en este caso UTF-8, que incluye casi todos los caracteres de todos los idiomas humanos
### \<title> \</title> - Elemento Título
#### Establece el título de tu página, que es el título que aparece en la pestaña o en la barra de título del navegador cuando la ágina es cargada
### \<body> \</body> - Elemento Body
#### Encierra todo el contenido que deseas mostrar a los usuarios web que visiten tu página, ya sea texto, imágenes, videos, juegos, pistas de audio reproducibles, y demás





# Imágenes


#### Como se mencionó anteriormente, incrusta una imagen en la página, en la posición que aparece. Esto es logrado a través del atributo "src", el cuál contien el path del archivo de imagen
#### También se incluye un atributo "alt", alternativo, que contiene un texto que debería describir la imagen y podría ser accedido por los usuarios que no pueden ver la imagen

# Marcado de Texto
#### Veremos algunos elementos HTML básicos para el marcado de texto
## Encabezados

#### Permiten especificar que ciertas partes del contenido son encabezados o subencabezados del contenido
<h1>Mi título principal</h1>
<h2>Mi título de nivel superior</h2>
<h3>Mi subtítulo</h3>
<h4>Mi sub-subtítulo</h4>

## Párrafos

#### Se utilizan para encerrar párrafos de texto
<p>Este es un simple parrafo</p>

## Listas

#### Mucho contenido web está dado por listas; y por esto HTML tiene elementos especiales para ellas
#### El marcado de elementos se realiza siempre en al menos dos elementos. El marcado de listas se realiza siempre en al menos dos elementos. Los dos tipos de listas más comunes son las listas ordenadas y las desordenadas:
#### __Desordenadas:__
#### Son aquellas en las que el orden de los items no es relevante \<ul> - unordered list
#### __Ordenadas:__
#### Son aquellas en las que el orden sí es relevante \<ul> - ordered list
<p>En Mozilla, somos una comunidad de</p>

<ul>
  <li>tecnólogos</li>
  <li>pensadores</li>
  <li>constructores</li>
</ul>

<p>trabajando juntos...</p>
