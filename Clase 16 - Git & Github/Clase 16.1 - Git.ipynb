# ¿Que es GIT?

#### Es un sistema de control de versiones gratuito y de código abierto, diseñado para manejar desde pequeños a grandes proyectos de manera rápida y eficaz
#### Es un sistema que registra los cambios realizados sobre un archivo o conjunto de archivos a lo largo del tiempo
#### Con GIT, podemos ir a versiones anteriores; muy útil para errores y para la organización



## Los 3 Estados

### 1er Estado - Working Directory
#### Comienzo del trabajo
### 2do Estado - Staging Area
#### Archivos listos
### 3er Estado - Repository
#### Registro de todos los archivos


## Configurando Git por primera vez: Tu Identidad
#### Lo primero que debemos hacer es establecer nuestro nombre de usuario y dirección de correo electrónico
#### Esto es importante porque los "commits" en Git utilizan esta información y es introducida de manera imutable en los commits que enviamos

##### git config --global user.name "Guido Soria"
##### git config --global user.email "g.sorianantes@gmail.com"

## ¿Que es un repositorio?
#### Es un espacio centralizado donde se almacena, oganiza, mantiene y difunde información
#### Será "la carpeta" donde guardaremos nuestro proyecto para más adelante compartirlo con el equipo a través de un repositorio en la nube (en internet, por ejemplo en Github)

## Git Init
#### Este comando se usa para crear un nuevo repositorio en Git
#### Nos crea un repositorio de manera local y lo hará en la carpeta en la cuál estamos parados
##### "git init" > Para inicializar un repositorio con la carpeta corriente
##### "git init nombre" > Para inicializar un repositorio creando una nueva carpeta

## Git Add
#### Ahora necesitamos agregar el o los archivos que queramos al "Staging Area"
##### git add "archivo" > Podemos pasar archivos indicando cuáles
##### git add . > Indicamos todos los archivos de la carpeta

## Git Commit
#### Una vez que nuestros archivos están en "Staging Area" debemos pasarlos a nuestro repositorio local y para eso debemos usar "git commit"
#### Esto "saca una foto" a nuestro repositorio. Si hicimos "commit" en un futuro siempre podremos volver a él

## Git Status
#### El comando "git status" nos permite ver el estado de nuestro repositorio: En que rama estamos, si hay archivos modificados sin pasar a stage, si hay cambios sin commitear, etc

## Git Log
#### Tenemos logs de los "commits" que hemos hecho

# Ramas
#### Una "rama" o "branch" es una versión separada del proyecto en la que podemos trabajar sin afectar la rama principal, normalmente llamada "master" o "main"
#### Las ramas nos permiten dessarrollar funcionalidades, corregir errores, o experimentar con ideas nuevas de manera aislada del resto del proyecto
#### Esto nos brinda: 
##### - Desarrollo Paralelo: Podemos trabajar en múltiples características o correcciones al mismo tiempo sin interferir con el código estable o trabajo de otros
##### - Aislamiento: Permiten probar y depurar cambios sin riesgo de romper el código existente
##### - Colaboración: Facilita el trabajo en equipo ya que cada desarrollador puede trabajar en su propia rama sin afectar a los demás. Al finalizar los cambios se pueden combinar (marge) en una rama común
##### - Control de veriones: Las ramas permiten gestionar diferentes versiones de una aplicación, como versiones de desarrollo, producción o experimentales


## git branch
#### Nos da un listado de las ramas que tenemos en nuestro proyecto
## git branch [nombre]
#### Creamos una nueva rama a partir del punto en el que estamos parados
## git checkout [nombre]
#### Nos sirve para movernos a otra rama


#### Los cambios que hagamos en la rama quedarán solo en esa rama hasta que hagamos un marge, si nos movemos de rama, los cambios que commiteamos en esa rama y no mergeamos desaparecerán:


## MERGE - git merge
#### Para hacer merge nos paramos en la rama "main" o la rama a la que queremos mergear nuestros cambios y utilizamos el comando "git merge [nombre]
