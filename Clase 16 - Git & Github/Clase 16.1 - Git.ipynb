# Que es GIT?

#### Es un sistema de control de versiones gratuito y de cdigo abierto, diseado para manejar desde pequeos a grandes proyectos de manera rpida y eficaz
#### Es un sistema que registra los cambios realizados sobre un archivo o conjunto de archivos a lo largo del tiempo
#### Con GIT, podemos ir a versiones anteriores; muy til para errores y para la organizacin



## Los 3 Estados

### 1er Estado - Working Directory
#### Comienzo del trabajo
### 2do Estado - Staging Area
#### Archivos listos
### 3er Estado - Repository
#### Registro de todos los archivos

## Que es un repositorio?
#### Es un espacio centralizado donde se almacena, oganiza, mantiene y difunde informacin
#### Ser "la carpeta" donde guardaremos nuestro proyecto para ms adelante compartirlo con el equipo a travs de un repositorio en la nube (en internet, por ejemplo en Github)

## Git Init
#### Este comando se usa para crear un nuevo repositorio en Git

##### "git init" > Para inicializar un repositorio con la carpeta corriente
##### "git init nombre" > Para inicializar un repositorio creando una nueva carpeta

# Ramas
#### Una "rama" o "branch" es una versin separada del proyecto en la que podemos trabajar sin afectar la rama principal, normalmente llamada "master" o "main"
#### Las ramas nos permiten dessarrollar funcionalidades, corregir errores, o experimentar con ideas nuevas de manera aislada del resto del proyecto
#### Esto nos brinda: 
##### - Desarrollo Paralelo: Podemos trabajar en mltiples caractersticas o correcciones al mismo tiempo sin interferir con el cdigo estable o trabajo de otros
##### - Aislamiento: Permiten probar y depurar cambios sin riesgo de romper el cdigo existente
##### - Colaboracin: Facilita el trabajo en equipo ya que cada desarrollador puede trabajar en su propia rama sin afectar a los dems. Al finalizar los cambios se pueden combinar (marge) en una rama comn
##### - Control de veriones: Las ramas permiten gestionar diferentes versiones de una aplicacin, como versiones de desarrollo, produccin o experimentales


## git branch

## git branch [nombre]

## git checkout [nombre]
#### Nos sirve para movernos a otra rama
